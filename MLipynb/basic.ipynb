```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#生成随机数据，图形
```

```python
x = np.random.uniform(0.0,1.0,50) #均匀分布
x
```

```python
y=np.random.normal(0.0,1.0,50)
y
```

```python
fig,ax = plt.subplots()
plt.scatter(x,y)
```

```python
fig,ax = plt.subplots()
plt.xlabel('Random Uniform')
plt.ylabel("Random Normal")
plt.scatter(x,y,marker='o',color='red')#散点图
```

```python
plt.subplot(121)#定义了一行，两列，第一张图
plt.scatter(x,y)
plt.subplot(122)#第二张图
plt.hist(y)
```
